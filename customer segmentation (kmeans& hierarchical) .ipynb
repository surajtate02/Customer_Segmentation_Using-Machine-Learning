#  Behavioral Segmentation Using K-Means Clustering ML Algorithm &  Hierarchical Clustering Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Customers.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df
```

```python
x = df.iloc[:,[3,4]].values
x
```

```python
plt.scatter(x[...,0],x[...,1])
plt.xlabel('Total Income')
plt.ylabel('spending score')
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
#within cluster some of square
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Cluster")
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
Y_kmeans = kmeans.fit_predict(x)
Y_kmeans
```

```python
x[Y_kmeans==0,0]
```

```python
x[Y_kmeans==0,1]
```

```python
plt.scatter(x[Y_kmeans==0,0],x[Y_kmeans==0,1],label='Standard')
plt.scatter(x[Y_kmeans==1,0],x[Y_kmeans==1,1],label='Careless')
plt.scatter(x[Y_kmeans==2,0],x[Y_kmeans==2,1],label='Target')
plt.scatter(x[Y_kmeans==3,0],x[Y_kmeans==3,1],label='Cluster 4')
plt.scatter(x[Y_kmeans==4,0],x[Y_kmeans==4,1],label='Sensible')

plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
df['Target'] = Y_kmeans
df
```

# Hierarchical Clustering Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Customers.csv")
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,[3,4]].values
X
```

```python
plt.scatter(X[...,0],X[...,1])
plt.xlabel("Total Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distances")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
Y_hc = hc.fit_predict(X)
Y_hc
```

```python
plt.scatter(X[Y_hc==0,0],X[Y_hc==0,1],label='Careful')
plt.scatter(X[Y_hc==1,0],X[Y_hc==1,1],label='Standard')
plt.scatter(X[Y_hc==2,0],X[Y_hc==2,1],label='Target')
plt.scatter(X[Y_hc==3,0],X[Y_hc==3,1],label='Careless')
plt.scatter(X[Y_hc==4,0],X[Y_hc==4,1],label='Sensible')

plt.legend()
plt.show()
```

```python
df['Target'] = Y_hc
df
```
